```python
from tqdm.notebook import tqdm
```

```python
fp_on_all = defaultdict(set)
run_drug_pairs = set(zip(df["run"], df["drug"]))
```

```python
tools = set(df["tool"])
```

```python
for (run, drug) in tqdm(run_drug_pairs):
    clfs = [df.at[(run, drug, t), "classification"] for t in tools]
    if all(c=="FP" for c in clfs):
        fp_on_all[run].add(drug)
```

```python
for i in range(1, len(set(df["drug"]))+1):
    n = sum(1 for v in fp_on_all.values() if len(v)>=i)
    print(f"{n} runs have FP for all tools, for {i} or more drugs")
```

```python
drug_class = {"aminoglycoside": {"amikacin", "capreomycin", "kanamycin"}, "fluoroquinolones": {"levofloxacin", "moxifloxacin", "ofloxacin"}}
drug_class_tmp = dict()
for cls, drugs in drug_class.items():
    for d in drugs:
        drug_class_tmp[d] = cls
drug_class = drug_class_tmp
```

```python
for run, drugs in fp_on_all.items():
    updated_drugs = {drug_class.get(d, d) for d in drugs}
    fp_on_all[run] = updated_drugs
```

```python
for i in range(1, len(set(df["drug"]))+1):
    n = sum(1 for v in fp_on_all.values() if len(v)==i)
    inh_rif = sum(1 for v in fp_on_all.values() if len(v)==i and ("rifampicin" in v and "isoniazid" in v))
    print(f"- {n} runs have FP for {i} drug(s), on all tools  - {inh_rif} are FP for RIF and INH")
```

Looking at significant sensitivity and specificity differences

```python
summary = pd.read_csv("../../results/tables/sn_sp/summary.illumina.csv")
summary.set_index(["Drug", "Tool"], drop=False, inplace=True)
summary
```

```python
from itertools import combinations
import re
from intervaltree import Interval
```

```python
rx = re.compile(r"(?P<lwr>\d+\.\d)-(?P<upr>\d+\.\d)%")
def extract_ci(s: str) -> tuple[float, float]:
    m = rx.search(s)
    if not m:
        raise ValueError(s)
    lwr = float(m.group("lwr"))
    upr = float(m.group("upr"))
    return lwr, upr
```

```python
drugs = set(summary["Drug"])
tools = set(summary["Tool"])
```

```python
results = []
for drug in drugs:
    for t1, t2 in combinations(tools, 2):
        t1_sn = summary.at[(drug, t1), "Sensitivity (95% CI)"]
        t1_sn_ci = extract_ci(t1_sn)
        t2_sn = summary.at[(drug, t2), "Sensitivity (95% CI)"]
        t2_sn_ci = extract_ci(t2_sn)
        
        sn_ci_overlaps = Interval(t1_sn_ci[0], t1_sn_ci[1] + 0.01).overlaps(t2_sn_ci[0], t2_sn_ci[1] + 0.01)
        if not sn_ci_overlaps:
            print(f"{drug} Sensitivity CI does not overlap for {t1} ({t1_sn_ci}) & {t2} ({t2_sn_ci})")
            results.append(["sensitivity", drug, t1, t1_sn_ci, t2, t2_sn_ci])
        
        t1_sp = summary.at[(drug, t1), "Specificity (95% CI)"]
        t1_sp_ci = extract_ci(t1_sp)
        t2_sp = summary.at[(drug, t2), "Specificity (95% CI)"]
        t2_sp_ci = extract_ci(t2_sp)
        
        sp_ci_overlaps = Interval(t1_sp_ci[0], t1_sp_ci[1] + 0.01).overlaps(t2_sp_ci[0], t2_sp_ci[1] + 0.01)
        if not sp_ci_overlaps:
            print(f"{drug} Specificity CI does not overlap for {t1} ({t1_sp_ci}) & {t2} ({t2_sp_ci})")
            results.append(["specificity", drug, t1, t1_sp_ci, t2, t2_sp_ci])
```

```python
print(pd.DataFrame(results, columns=["Metric", "Drug", "Tool1", "Tool1 CI", "Tool2", "Tool2 CI"]).to_markdown(index=False))
```
